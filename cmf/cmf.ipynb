```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('cmf_data_1.csv')
```

```python
df.head()
```

```python
#naming the columns
df.columns = ['T', 'muB', 'muS', 'muQ', 'nB', 'nS', 'nQ',  'E', 'P', 's', 'sigma_MF', 'zeta_MF', 'delta_MF', 'omega_MF', 'phi_MF', 'rho_MF', 'phi_order_F', 'nB_noPol', 'Quark_nB', 'Octet_nB', 'Decuplet_nB']
df.head()
```

```python
#plotting pressure vs energy columns
plt.figure(figsize=(10, 6))
plt.plot(df['E'], df['P'], label='EoS:Pressure vs Energy', color='blue')

plt.xlabel('Energy (E)')
plt.ylabel('Pressure (P)')
plt.title('Pressure vs Energy')
plt.legend()
plt.grid()

#saving the plot
plt.savefig('pressure_vs_energy1.png')
plt.show()
```

```python
#plotting  mean fields vs chemical potentials
plt.figure(figsize=(10, 6))
plt.plot(df['muB'], df['sigma_MF'], label='Sigma Mean Field vs Baryon Chemical Potential', color='red')
plt.plot(df['muB'], df['omega_MF'], label='Omega Mean Field vs Baryon Chemical Potential', color='green')         
plt.xlabel('Baryon Chemical Potential (muB)')    
plt.ylabel('Mean Fields')
plt.title('Mean Fields vs Baryon Chemical Potential')
plt.legend()
plt.tight_layout()
plt.grid()  
plt.savefig('mean_fields_vs_chemical_potentials1.png') 
plt.show() 
```

```python
# 3D plot sigmam, muB,muQ
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') 
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('Sigma Mean Field')
ax.set_ylabel('Baryon Chemical Potential (muB)')
ax.set_zlabel('Baryon Chemical Potential (muQ)')
ax.set_title('3D Plot of Sigma Mean Field vs Baryon Chemical Potentials')
plt.show()  
```
