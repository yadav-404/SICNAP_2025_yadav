## Task - 1 :Simple harmonic motion

```python


# Parameters
k = 1.0 # Spring constant (N/m)
A = 1.0  # Amplitude (m)


# User chooses mass
print("Choose a mass for the mass-spring system: 1 kg, 2 kg, 3 kg, or 4 kg.")
mass_choice = int(input("Choose mass"))

if mass_choice == 1:
    m = 1 # kg
    print("You chose mass = 1 kg.")
elif mass_choice == 2:
    m = 2  # kg
    print("You chose mass = 2 kg.")
elif mass_choice == 3:
    m = 3 # kg
    print("You chose mass = 3 kg.")
else:
    print("OOPs Invalid choice, using default mass = 1 kg.")
    m = 1

# Calculate angular frequency
omega = np.sqrt(k / m)

# Time settings
t_max = 10  # seconds
dt = 0.1  # time step
time_steps = int(t_max / dt)

# Arrays to store time and displacement
t_values = []
x_values = []

# Calculate displacement at each time step
for i in range(time_steps):
    t = round(i * dt,2)
    x = round(A * np.sin(omega * t),2)
    t_values.append(t)
    x_values.append(x)
print("Displacement values:", x_values)
print("Time values:", t_values)


# Plot displacement vs time

plt.plot(t_values, x_values, label=f"Mass = {m} kg")
plt.title("Simple Harmonic Motion of Mass-Spring System")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.grid()
plt.legend()
plt.axhline(0, color="black", lw=1, ls="--")  # Add horizontal line at y=0
plt.xlim(0, t_max)
plt.show()

```

## Task 1.2 Projectile motion

```python
#parameters for projectile motion
theta_deg = 45  # launch angle
g = 10  # gravity


print("Enter the initila velocity")
v0 = float(input("Initial speed (m/s): "))
print(f"Initial speed: {v0} m/s, Launch angle: {theta_deg} degrees, Gravity: {g} m/s")

# Time settings
dt = 0.1  # time step
t = 0

# Lists to store trajectory
x_values = []
y_values = []

# Simulation loop
while True:
    x = round(v0 * np.cos(theta_deg) * t,2)
    y = round(v0 * np.sin(theta_deg) * t - 0.5 * g * t**2,2)
    
    print(f"At time {t:.2f} s: x = {x:.2f} m, y = {y:.2f} m")
    
    # Stop if projectile hits the ground
    if y < 0:
        break
    else:
        x_values.append(x)
        y_values.append(y)
    
    t += dt

# Plotting
plt.plot(x_values, y_values, marker="o")
plt.title("Projectile Motion Trajectory")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid(True)
plt.ylim(0, max(y_values)+1)
plt.xlim(0, max(x_values)+1)
plt.show()
```

##  Task 1.3 Planetary motion

```python


# Parameters
R = 1.0  # orbital radius (AU)
T = 365  # orbital period (days)
omega = 2 * np.pi / T  # angular velocity

# Time settings
dt = 1  # time step 1 day
t = 0

# Lists to store positions
x_values = []
y_values = []

# Simulation loop: run until one full orbit is completed
while t <= T:
    x =  round(R * np.cos(omega * t),2)
    y = round(R * np.sin(omega * t),2)
    x_values.append(x)
    y_values.append(y)
    
    t += dt

# Plot orbit
plt.plot(x_values, y_values, marker="o", markersize=1)
plt.title("Planetary Circular Orbit")
plt.xlabel("x (AU)")
plt.ylabel("y (AU)")
plt.grid(True)
plt.show()
```

## Matrix multiplication

```python
# Matrix multiplication example using numpy

# Define two matrices A and B
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Multiply using @ operator
product_at = A @ B

# Multiply using numpy.dot() function
product_dot = np.dot(A, B)

# Display matrices
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nProduct using @ operator:")
print(product_at)
print("\nProduct using numpy.dot():")
print(product_dot)
```

Block sliding down the inclined plane

```python
# Block sliding down an inclined plane

# Parameters
g = 9.81  # gravity (m/s^2)
theta = 180 * np.pi / 180  # incline angle in degrees

a = g * np.sin(theta)  # acceleration along incline

# Time steps
t = np.linspace(0, 50, 20)
#print(t)
# Position and velocity
s = 0.5 * a * t**2
v = a * t

s_values = []
v_values = []
for i in t:
    s= 0.5 * a * i**2  # Calculate position at time i
    v = a * i  # Calculate velocity at time i

    #s[i] = round(s[i], 2)  # Round position to 2 decimal places
    s_values.append(s)
    v_values.append(v)

print("Position values:", s_values)
# Plot position
plt.figure(figsize=(8, 5))
plt.plot(t, s_values, label='Position (m)')
# plt.plot(t, v_values, label='Velocity (m/s)')
plt.title(f'Block Sliding on {theta} Inclined Plane')
plt.xlabel('Time (s)')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show()
```

# Reusable Plotting Function

Defining function

```python
# Scatter plot

def scatter_plot(df, x_col, y_col):

    # Check if columns exist
    if x_col not in df.columns or y_col not in df.columns:
        raise ValueError(f"One or both columns "{x_col}", "{y_col}" do not exist in the DataFrame.")
    
    plt.figure(figsize=(8,6))
    plt.scatter(df[x_col], df[y_col], color="teal")
    plt.title(f"{y_col} vs {x_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.show()
```
